```python
df_cheatsheet = pd.read_json("labelstudio_cheatsheet.json")
```

```python
df_tobias = pd.read_json("tobias_svar_alle.json")
df_samir = pd.read_json("samir_svar_alle.json")
df_jack = pd.read_json("jack_svar_alle.json")
```

```python
tobias_tekstsvar = list()

for i in range(len(df_tobias)):
    answers = df_tobias["annotations"].iloc[i][0]["result"]

    if len(answers) > 2:
        print(answers[2])
        tobias_tekstsvar.append(answers[2]["value"]["text"][0])
    else:
        print("No answer")
        tobias_tekstsvar.append("No answer")
    
```

```python
samir_tekstsvar = list()

for i in range(len(df_samir)):
    answers = df_samir["annotations"].iloc[i][0]["result"]

    if len(answers) > 2:
        print(answers[2])
        samir_tekstsvar.append(answers[2]["value"]["text"][0])
    else:
        print("No answer")
        samir_tekstsvar.append("No answer")
```

```python
jack_tekstsvar = list()

for i in range(len(df_samir)):
    answers = df_jack["annotations"].iloc[i][0]["result"]

    if len(answers) > 2:
        print(answers[2])
        jack_tekstsvar.append(answers[2]["value"]["text"][0])
    else:
        print("No answer")
        jack_tekstsvar.append("No answer")
```

```python
df_answers = pd.DataFrame([
    {
        "id":df_tobias.iloc[i]["inner_id"],
        "tobias_answer_q1":df_tobias["annotations"].iloc[i][0]["result"][0]["value"]["choices"],
        "tobias_answer_q2":df_tobias["annotations"].iloc[i][0]["result"][1]["value"]["choices"],
        "samir_answer_q1":df_samir["annotations"].iloc[i][0]["result"][0]["value"]["choices"],
        "samir_answer_q2":df_samir["annotations"].iloc[i][0]["result"][1]["value"]["choices"],
        "jack_answer_q1":df_jack["annotations"].iloc[i][0]["result"][0]["value"]["choices"],
        "jack_answer_q2":df_jack["annotations"].iloc[i][0]["result"][1]["value"]["choices"],        
    } for i in range(len(df_tobias))
])
```

```python
df_results  = pd.concat((df_cheatsheet,df_answers), axis=1)
```

```python
df_results["tobias_tekstsvar"] = tobias_tekstsvar
df_results["samir_tekstsvar"] = samir_tekstsvar
df_results["jack_tekstsvar"] = jack_tekstsvar
```
